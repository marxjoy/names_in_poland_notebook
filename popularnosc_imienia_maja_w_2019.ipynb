# Popularność imienia MAJA w 2019 roku

```python
https://dane.gov.pl/dataset/219,imiona-nadawane-dzieciom-w-polsce
```

```python
df = pd.read_csv('imiona_żeńskie_województwa_2019_imię_pierwsze.csv')
```

# W 2019 roku urodziło sie  w Polsce 186 tysięcy 405 dziewczynek

```python
df['LICZBA_WYSTĄPIEŃ'].sum()
```

# Nadano im 509 różnych pierwszych imion żeńskich

```python
len(df['IMIĘ_PIERWSZE'].unique())
```

# Najpopularniejsze imiona to: Zuzanna, Julia, Zofia, Maja, Hanna, Lena

## Urodziło się 7435 Maj.

```python
df.groupby('IMIĘ_PIERWSZE').sum().sort_values('LICZBA_WYSTĄPIEŃ', ascending=False).head(50).drop(columns=['WOJ'])
```

# Najmniej popularne imiona:

```python
df.groupby('IMIĘ_PIERWSZE').sum().sort_values('LICZBA_WYSTĄPIEŃ', ascending=False).tail(60).drop(columns=['WOJ'])
```

```python
maja_woj = df[df['IMIĘ_PIERWSZE'] == 'MAJA']
```

```python
maja_woj['procent'] = maja_woj['LICZBA_WYSTĄPIEŃ'] / 7435 * 100 #
```

# Spośród Mai urodzonych w 2019 roku, 14,5% urodziło się w mazowieckim, najmniej w Opolskim.

```python
maja_woj.sort_values('procent', ascending=False).drop(columns=['WOJ', 'IMIĘ_PIERWSZE', 'PŁEĆ'])
```

```python
urodzenia = df.groupby('WOJEWÓDZTWO').sum()
```

```python
woj_maja = maja_woj.set_index('WOJEWÓDZTWO')
```

```python
woj_maja['LICZBA_MAJ'] = woj_maja['LICZBA_WYSTĄPIEŃ']
```

```python
woj_maja.drop(columns=['WOJ', 'IMIĘ_PIERWSZE', 'PŁEĆ', 'LICZBA_WYSTĄPIEŃ', 'procent'], inplace=True)
```

```python
result = pd.concat([woj_maja, urodzenia], axis=1)
```

```python
result.drop(columns=['WOJ'], inplace=True)
```

```python
result['procent'] = result['LICZBA_MAJ'] * 100 / result['LICZBA_WYSTĄPIEŃ']
```

```python
result.rename(columns={'LICZBA_WYSTĄPIEŃ': 'LICZBA_URODZEŃ'}, inplace=True)
```

# Tyle procent spośród wszystkich imion było Majami w każdym województwie

```python
result.sort_values('procent', ascending=False)
```

```python
zestaw_imion = df.groupby(['WOJEWÓDZTWO']).count()
```

```python
zestaw_imion['unikalne'] = zestaw_imion['IMIĘ_PIERWSZE']
```

```python
zestaw_imion.drop(columns=['WOJ','IMIĘ_PIERWSZE','PŁEĆ','LICZBA_WYSTĄPIEŃ'], inplace=True)
```

# Tyle różnych imion żeńskich nadano w 2019 w różnych województwach

```python
res = pd.concat([urodzenia, zestaw_imion], axis=1)
```

```python
res.rename(columns={'unikalne': 'unikalne_imiona'}, inplace=True)
```

```python
res.sort_values('unikalne_imiona', ascending=False).drop(columns=['WOJ'])
```

```python
df.drop(columns=['WOJ', 'PŁEĆ'], inplace=True)
```

```python
woje = list(df['WOJEWÓDZTWO'].unique())
```

# Pierwsze 6 najpopularniejszych imion żeńskich nadanych w 2019 roku z podziałem na województwa

```python
for w in woje:
    print(df[df['WOJEWÓDZTWO'] == w][0:6].reset_index().drop(columns=['index']))
    print('===================================')
```
