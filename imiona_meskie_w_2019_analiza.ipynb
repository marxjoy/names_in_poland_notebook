```python
df = pd.read_csv('imiona_męskie_województwa_2019_imię_pierwsze.csv')
```

# Nadano im 508 różnych pierwszych imion męskich

```python
len(df['IMIĘ_PIERWSZE'].unique())
```

# Najpopularniejsze imiona to: Antoni, Jan, Jakub, Aleksander, Szymon



```python
df.groupby('IMIĘ_PIERWSZE').sum().sort_values('LICZBA_WYSTĄPIEŃ', ascending=False).head(50).drop(columns=['WOJ'])
```

# Najmniej popularne imiona:

```python
df.groupby('IMIĘ_PIERWSZE').sum().sort_values('LICZBA_WYSTĄPIEŃ', ascending=False).tail(60).drop(columns=['WOJ'])
```

```python
urodzenia = df.groupby('WOJEWÓDZTWO').sum()
```

```python
zestaw_imion = df.groupby(['WOJEWÓDZTWO']).count()
```

```python
zestaw_imion['unikalne'] = zestaw_imion['IMIĘ_PIERWSZE']
```

```python
zestaw_imion.drop(columns=['WOJ','IMIĘ_PIERWSZE','PŁEĆ','LICZBA_WYSTĄPIEŃ'], inplace=True)
```

# Tyle różnych imion męskich nadano w 2019 w poszczólnych województwach

```python
res = pd.concat([urodzenia, zestaw_imion], axis=1)
```

```python
res.rename(columns={'unikalne': 'unikalne_imiona'}, inplace=True)
```

```python
res.sort_values('unikalne_imiona', ascending=False).drop(columns=['WOJ'])
```

```python
df.drop(columns=['WOJ', 'PŁEĆ'], inplace=True)
```

```python
woje = list(df['WOJEWÓDZTWO'].unique())
```

# Pierwsze 6 najpopularniejszych imion męskich nadanych w 2019 roku z podziałem na województwa

```python
for w in woje:
    print(df[df['WOJEWÓDZTWO'] == w][0:6].reset_index().drop(columns=['index']))
    print('===================================')
```
